```julia
# Length in 1D of our 3D matrix
N=3

println("1D size: ",N," # of units N^3: ",N^3)

# Zero our dense Hamiltonian matrix
H=zeros(Int,N^3,N^3)

# Cross ref from x,y,z tuple to matrix location
#  - nb, intent is to expand this to a Arrowhead forming version!
#    Once we figure out the arithmatic :)
function matrixloc(x,y,z)
    return ((x+N)%N + ((y+N)%N)*N + ((z+N)%N)*N*N)+1
end

# OK; loop over all connections between all possible locations.
for x=0:N-1
    for y=0:N-1
        for z=0:N-1
            print(x," ",y," ",z," ", matrixloc(x,y,z)," :\n") 
            
            #H[matrixloc ; matrixloc] = 1
            
            for dx=-1:2:1 # TODO: this loops over all 8 edges in 3x3x3 supercell!
                for dy=-1:2:1
                    for dz=-1:2:1
                        print(matrixloc(x,y,z),"  ",matrixloc(x+dx,y+dy,z+dz),"\n")
                        H[matrixloc(x,y,z),matrixloc(x+dx,y+dy,z+dz)]=1 # Set value in Hamiltonian H
                    end
                end
            end
            println()
        end
    end
end
```
